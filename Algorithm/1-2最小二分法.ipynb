```python
import numpy as np;
import tensorflow as tf;
import matplotlib.pyplot as plt;
```

```python
def line(points):
    #求x，y得平均值
    averageXY=np.mean(points,1);
    #获取x,y数据
    x=points[0];
    y=points[1];
    #求x，y与平均值得差
    avX=x-averageXY[0,0];
    avY=y-averageXY[1,0];
    #求差值平方
    squX=avX**2;
    squY=avY**2;
    #偏方差
    squXY=avX*avY;
    #求平方和
    _x=np.sum(squX);
    _y=np.sum(squY);
    _xy=np.sum(squXY);
    
    a=_xy/_x;#回归系数A
    b=averageXY[1,0]-_xy*averageXY[0,0]/_x;#回归系数b
    
    r=_xy/(np.sqrt(_x*_y));#回归系数
    
    print("a="+str(a)+" b="+str(b)+" r="+str(r));
    
    y_data=x*a+b;
    #绘制图
    plt.figure();
    #实际值
    plt.scatter(x,y);
    #与值
    plt.plot(x,y_data,"r-",lw=5);
    plt.show();

if __name__=="__main__":
    #生成线性随机x数据,数据-5到5，总共200个
    x_data=np.linspace(5,15,200)[:,np.newaxis];
    #生成噪声
    noise=np.random.normal(0,0.5,x_data.shape);
    #生成y数据
    y_data=3*x_data+noise;
    #数据集
    dataSet=np.array([x_data,y_data]);
    
    line(dataSet);
```
